```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# create a grid of x y axis point and evaluate a trig function on that point and plot gradient.

xx  = np.linspace(-3, 3, 51) #creating a set of point that goes from -3 to 3 in 51 steps
X, Y = np.meshgrid(xx, xx)

Z = np.sin(X*np.cos(Y))
gx, gy = np.gradient(Z)

plt.contourf(xx, xx, Z, 40)
plt.quiver(xx, xx, gx, gy);
```

```python
xx  = np.linspace(-3, 3, 51) #creating a set of point that goes from -3 to 3 in 51 steps
X, Y = np.meshgrid(xx, xx)

f1 = 4
f2 = .8

Z = np.sin(f1*X*np.cos(f2*Y))
gx, gy = np.gradient(Z)

plt.contourf(xx, xx, Z, 40)
plt.quiver(xx, xx, gx, gy);
```

```python
xx  = np.linspace(-3, 3, 51) #creating a set of point that goes from -3 to 3 in 51 steps
X, Y = np.meshgrid(xx, xx)

f1 = 4
f2 = .8
p1 = .6
p2 = -.77

Z = np.sin(f1*X*np.cos(f2*Y+p2)+p1)
gx, gy = np.gradient(Z)

plt.contourf(xx, xx, Z, 40 , cmap = 'Purples')
plt.quiver(xx, xx, gx, gy);
```
